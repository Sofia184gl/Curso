```python
import networkx as nx
import plotly.express as px 
import matplotlib.pyplot as plt
```

```python
data=px.data.gapminder()
data.shape 
```

```python
data.head()
```

```python
data['nation']='Country'
data=nx.from_pandas_edgelist(data,source='nation', target ='country', edge_attr=True,)
```

```python
import warnings

plt.rcParams['figure.figsize'] = (22,20)
plt.style.use('fivethirtyeight')
plt.title('Countries in the World', fontsize =30)
plt.grid()
plt.axis('off')

pos =nx.spring_layout(data)

#drawing nodes
nx.draw_networkx_nodes(data, pos, node_size =1200, node_color = 'pink')

#labels
nx.draw_networkx_nodes(data, pos, font_size =20, font_family = 'sans-serif')

#drawing edges
nx.draw_networkx_nodes(data, pos, width=6, alpha=0.1, edge_color = 'brown')

plt.show()
```
